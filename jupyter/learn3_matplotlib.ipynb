```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

```python
plt.plot(x,y)
plt.xlabel("Angle")
plt.ylabel("sine value")
plt.title("Sin Wave")
plt.show()
```

```python
plt.plot(x,z)
plt.show()
```

```python
plt.plot(x,y,'g-')
plt.plot(x,z,'r--')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
lang = ['English','Tamil','French']
people = [100,80,60]
ax.bar(lang,people)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
lang = ['English','Tamil','French']
people = [100,80,60]
ax.pie(people,labels=lang,autopct='%1.1f')
plt.show()
```

```python
x = np.linspace(0,10,50)
y = np.sin(x)
z = np.cos(x)

fig2 = plt.figure()
ax1 = fig2.add_axes([0,0,1,1])
ax1.scatter(x,y,color = 'r')
ax1.scatter(x,z,color = 'g')
plt.show()
```

```python
fig3 = plt.figure()
ax = plt.axes(projection = '3d')
z = 20*np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z)
plt.show()
```
